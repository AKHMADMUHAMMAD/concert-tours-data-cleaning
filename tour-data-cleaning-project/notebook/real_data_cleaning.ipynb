```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/my_file (1).csv')
df.head(20)
```

```python
df = df.drop('Peak',axis=1)
df = df.drop('All Time Peak',axis=1)
df
```

```python
df.value_counts()
```

```python
# Faraz: Adjusted gross 3-ustun, Average gross 4-ustun (nomlarini bilmagan holda)
# Lekin nomini bilish yaxshiroq. Eng ishonchli:
df.iloc[:, [1,2,7]] = df.iloc[:,[1,2,7]].apply(lambda x: x.replace(r'[\$,eb\[\]\(\)]', '', regex=True).astype(float))
```

```python
df.columns = df.columns.str.replace(r'[^a-zA-Z0-9\s&]', '', regex=True)
```

```python
print(df.columns)
```

```python
df.columns = (
    df.columns
    .str.replace(r'\s+', ' ', regex=True)  # 2+ space → 1 ta space
    .str.strip()
)
```

```python
print(df.columns.tolist())
```

```python
df[['Actual gross','Adjusted gross in 2022 dollars','Average gross']] = \
df[['Actual gross','Adjusted gross in 2022 dollars','Average gross']].astype(int)
```

```python
df['Tour title']
```

```python
df['Tour title'] = (
    df['Tour title']
    .str.replace(r'\[.*?\]', '', regex=True)
    .str.replace(r'[^a-zA-Z0-9\s&]', '', regex=True)
    .str.strip()
)
```

```python
df = df.drop('Ref',axis=1)
df
```

```python
df['Years']
```

```python
# year(s) utsuni ustida tozalash va togirlash ishlarini ol;ib borish
# 1 qadam: belginio normalizatsiya qilish

df['Years'] = df['Years'].str.replace(r'[–—−]', '-', regex=True)

# 2-qadam split qilish

years =df['Years'].str.split('-',expand=True)

# 3-qadam yangi ustunlar yaratish

df['start_year']= years[0]
df['end_year'] = years[1]

# 4- qadam bosh end_yearni toldirish

df['end_year'] = df['end_year'].fillna(df['start_year'])

# 5-qadam INT ga otkazish

df['start_year'] = df['start_year'].astype(int)
df['end_year']= df['end_year'].astype(int)

```

```python
df = df.drop('Years', axis=1)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
